# Walmart Weekly Sales Forecasting Project - Data Preprocessing

Veri birleştirmesi ve eksik değerlerin doldurulmasını içeren notebook.

```python
import pandas as pd

train = pd.read_csv("../data/train.csv")
test = pd.read_csv("../data/test.csv")
stores = pd.read_csv("../data/stores.csv")
features = pd.read_csv("../data/features.csv")

# Önce stores, sonra features ile left join işlemi
data = pd.merge(train, stores, on="Store", how="left")
data = pd.merge(data, features, on=["Store", "Date"], how="left")

# print(data["IsHoliday_x"].equals(data["IsHoliday_y"]))
# İki IsHoliday değeri de aynı olduğundan birini kaldırıyoruz.
data = data.drop(columns="IsHoliday_y")
data = data.rename(columns={"IsHoliday_x": "IsHoliday"})

# print("Final veri seti için eksik değer kontrolü:")
# print(data.isnull().sum())
# Sadece Markdown sütunlarında eksikler var. NaN değerlerini sıfır hale getiriyorum.
markdown_columns = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
data[markdown_columns] = data[markdown_columns].fillna(0)

# Eksik değer kontrolü
# print(data.isnull().sum())
# Eksik değer kalmadı.
```
